Initialize Data

```
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy
%matplotlib inline

from google.colab import files
uploaded = files.upload()
```

```
#Red data from csv file for training and validation data
TrainingSet = numpy.genfromtxt("G3TrainData.csv", delimiter=",", skip_header=True)
ValidationSet = numpy.genfromtxt("G3ValidateData.csv", delimiter=",", skip_header=True)

# split into input (X) and output (Y) variables
X1 = TrainingSet[:,0:1]
Y1 = TrainingSet[:,1:2]

X2 = ValidationSet[:,0:1]
Y2 = ValidationSet[:,1:2]

plt.plot (X1,Y1)
plt.plot (X2,Y2)

plt.xlabel('X-Train, X-Validate')
plt.ylabel('Y-Train, Y-Validate')
```

Setup Model

```
# create model
model = Sequential()
model.add(Dense(50, input_dim=1, init='uniform', activation='relu'))
model.add(Dense(100, init='uniform', activation='relu'))
model.add(Dense(100, init='uniform', activation='relu'))
model.add(Dense(100, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='linear'))


# Compile model
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X1, Y1, epochs=2000, batch_size=100,  verbose=2)

# Calculate predictions
YP1 = model.predict(X1)
YP2 = model.predict(X2)

# Save predictions
numpy.savetxt("trainresults.csv", YP1, delimiter=",")
numpy.savetxt("valresults.csv", YP2, delimiter=",")
```

```
plt.plot(X2,Y2)
plt.plot(X2,YP2)
plt.xlabel('X-Validate-Actual, X-Validate-Predicted')
plt.ylabel('Y-Validate-Actual, Y-Validate-Predicted')
```

Plot Results

```
#Plot actual vs predition for training set
TestResults = numpy.genfromtxt("trainresults.csv", delimiter=",")
plt.plot(Y1,TestResults,'ro')
plt.title('Training Set')
plt.xlabel('Actual')
plt.ylabel('Predicted')

#Compute R-Square value for training set
TestR2Value = r2_score(Y1,TestResults)
print("Training Set R-Square=", TestR2Value)
```

```
#Plot actual vs predition for validation set
ValResults = numpy.genfromtxt("valresults.csv", delimiter=",")
plt.plot(Y2,ValResults,'ro')
plt.title('Validation Set')
plt.xlabel('Actual')
plt.ylabel('Predicted')

#Compute R-Square value for validation set
ValR2Value = r2_score(Y2,ValResults)
print("Validation Set R-Square=",ValR2Value)
```
